1: Data Exploration and Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv("/content/sample_data/TASK-ML-INTERN.csv")

# Print the available columns to verify the correct column name
print(data.columns)

# Replace 'DON_concentration' with the actual column name if it's different
# For example, if the actual column name is 'DON_conc', change the line below to:
# X = data.drop(columns=['DON_conc'])
X = data.drop(columns=["vomitoxin_ppb"])  # Features
y = data["vomitoxin_ppb"]  # Target

# Check for missing values and outliers
print(data.isnull().sum())
print(data.describe())

# Handle missing values (e.g., fill with mean)
# Select only numeric columns for filling missing values
numeric_cols = X.select_dtypes(include=np.number).columns
X[numeric_cols] = X[numeric_cols].fillna(X[numeric_cols].mean())

# Standardize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X[numeric_cols]) # Scale only numeric features

# Visualize average reflectance
plt.plot(numeric_cols, X_scaled.mean(axis=0)) # Use numeric columns for plotting
plt.xlabel("Wavelength Bands")
plt.ylabel("Average Reflectance")
plt.title("Average Spectral Reflectance")
plt.show()

# Heatmap
sns.heatmap(X_scaled[:50], cmap="viridis")  # First 50 samples
plt.show()
```

Dimensionality Reduction

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=0.95)  # Keep 95% variance
X_pca = pca.fit_transform(X_scaled)

# Explained variance
print("Explained variance ratio:", pca.explained_variance_ratio_.cumsum())

# Visualize
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap="viridis")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA: First Two Components")
plt.colorbar(label="DON Concentration")
plt.show()
```

#3. Model Training

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.20, random_state=42)

# Define model and parameters
rf = RandomForestRegressor(random_state=42)
param_grid = {"n_estimators": [100, 200], "max_depth": [10, 20, None]}

# Grid search
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring="neg_mean_squared_error")
grid_search.fit(X_train, y_train)

# Best model
best_rf = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)
```
